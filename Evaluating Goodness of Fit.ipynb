```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
```

```python
weather=pd.read_csv("weatherHistory.csv")
weather.head(3)
```

```python
weather["diff_temp"]=weather["Temperature (C)"]-weather["Apparent Temperature (C)"]

Y=weather["diff_temp"]
X=weather[["Humidity","Wind Speed (km/h)"]]

lrm=linear_model.LinearRegression()
lrm.fit(X,Y)
print("Coefficient:", lrm.coef_)
print("Bias:", lrm.intercept_)
```

```python
Regresyon formülasyonu:

Y = -2.438 + 3.029*x1 + 0.119*x2
```

```python
X=sm.add_constant(X)

result=sm.OLS(Y,X).fit()
result.summary()
```

R-square değeri ile adjusted R-square değeri aynı olmakla birlikte 0.288'dir. Bu demek oluyorki, tüm katsayılar hedef değişkenin varyansını %28.8 açıklamaktadır, yani hedef değişkeni %28.8 oranında açıkamakta kalan %71.2 oranında hedef değişken hakkında bilgi içermemektedir.Dolayısıyla varyansın büyük olmasını bekleriz, model yeterli değildir.R-square ile adjustedR-square değerin fazla olması hedef değişkeni açıklamada fazla bilgi içerdiğini söyler.

```python
weather["inter_hum_wind"]=weather["Humidity"]*weather["Wind Speed (km/h)"]

Y1=weather["diff_temp"]
X1=weather[["Humidity","Wind Speed (km/h)","inter_hum_wind"]]

lrm1=linear_model.LinearRegression()
lrm1.fit(X1,Y1)

print("Coefficient", lrm1.coef_)
print("Bias", lrm1.intercept_)
```

```python
Humidity ile wind speed ilişkisini özellik olarak modele eklediğimizde çıkan formül:
    
Y1 = -0.084 + -0.177*X11 -0.090*X12 + 0.297*X13
```

```python
X1=sm.add_constant(X1)

result1=sm.OLS(Y1,X1).fit()
result1.summary()
```

Nem ile rüzgarın ilişkisini modele eklediğimizde R-square ile adjusted R-square değeri %34.1 oranına ulaşarak artmıştır; yani önceden hedef değişkende açıklayamadığı bilgilerin bir kısmını açıklayabilmiştir; dolayısıyla yeni özelliğin eklenmesi modelin tahmin gücünü arttırdığından modelin içinde bu özelliği tutabiliriz.

```python
Y2=weather["diff_temp"]
X2=weather[["Humidity","Wind Speed (km/h)","Visibility (km)"]]
           
lrm2=linear_model.LinearRegression()
lrm2.fit(X2,Y2)
           
print("Coefficient", lrm2.coef_)
print("Bias", lrm2.intercept_)
```

```python
Visibility değişkeni özellik olarak modele eklendiğinde oluşan formül:
    
Y2 = -1.575 + 2.606*X21 + 0.119*X22 - -0.053*X23
```

```python
X2=sm.add_constant(X2)

result=sm.OLS(Y2,X2).fit()
result.summary()
```

Visibility değişkeni özellik olarak modele eklenince; R-square değeri düşerek %30.4, adjusted R-square değeri ise %30.3 olmuştur. Bu demek oluyor ki ilk modele göre visibility özelliği hedef değişkeni açıklamada daha fazla bilgiye sahiptir ancak; humidity ile wind ilişkisini modele eklediğimiz 2. örneğe nazaran daha az R-square değerine sahiptir, dolayısıyla humidity ile wind ilişkisinin modelde yer alması visibilitye göre daha fazla tahmin yapmamızı sağlar.

```python
train=pd.read_csv("train.csv")
train.head(3)
```

```python
# Missing values çok fazla olduğu için direkt silebiliriz.

train.drop("LotFrontage",axis=1,inplace=True)
train.drop("Alley",axis=1,inplace=True)
train.drop("Id",axis=1,inplace=True)
train.drop("FireplaceQu",axis=1,inplace=True)
train.drop("PoolQC",axis=1,inplace=True)
train.drop("Fence",axis=1,inplace=True)
train.drop("MiscFeature",axis=1,inplace=True)
```

```python
train["GarageCond"].fillna(train["GarageCond"].mode()[0],inplace=True)
train["MasVnrArea"].fillna(train["MasVnrArea"].mode()[0],inplace=True)
train["GarageQual"].fillna(train["GarageQual"].mode()[0], inplace=True)
train["GarageFinish"].fillna(train["GarageFinish"].mode()[0], inplace=True)
train["GarageYrBlt"].fillna(train["GarageYrBlt"].mode()[0], inplace=True)
train["GarageType"].fillna(train["GarageType"].mode()[0], inplace=True)
train["Electrical"].fillna(train["Electrical"].mode()[0], inplace=True)
train["BsmtFinType2"].fillna(train["BsmtFinType2"].mode()[0], inplace=True)
train["BsmtFinType1"].fillna(train["BsmtFinType1"].mode()[0], inplace=True)
train["BsmtQual"].fillna(train["BsmtQual"].mode()[0], inplace=True)
train["BsmtCond"].fillna(train["BsmtCond"].mode()[0], inplace=True)
train["BsmtExposure"].fillna(train["BsmtExposure"].mode()[0], inplace=True)
train["MasVnrType"].fillna(train["MasVnrType"].mode()[0], inplace=True)
```

```python
train["MiscVal"].replace(0,"none",inplace=True)
train["PoolArea"].replace(0,"none",inplace=True)
train["ScreenPorch"].replace(0,"none",inplace=True)
train["3SsnPorch"].replace(0,"none",inplace=True)
train["EnclosedPorch"].replace(0,"none",inplace=True)
train["OpenPorchSF"].replace(0,"none",inplace=True)
train["WoodDeckSF"].replace(0,"none",inplace=True)
train["LowQualFinSF"].replace(0,"none",inplace=True)
train["BsmtFinSF2"].replace(0,"none",inplace=True)
```

```python
def paved(PavedDrive):
    if PavedDrive=="Y":
        return 3
    elif PavedDrive=="P":
        return 2
    elif PavedDrive=="N":
        return 1

train["PavedDrive_new"]=train["PavedDrive"].apply(paved)
```

```python
def garagequal(GarageQual):
    if GarageQual=="Ex":
        return 5
    elif GarageQual=="Gd":
        return 4
    elif GarageQual=="TA":
        return 3
    elif GarageQual=="Fa":
        return 2
    elif GarageQual=="Po":
        return 1
    
train["GarageQual_new"]=train["GarageQual"].apply(garagequal)
```

```python
def garagefinish(GarageFinish):
    if GarageFinish=="Fin":
        return 3
    elif GarageFinish=="RFn":
        return 2
    elif GarageFinish=="Unf":
        return 1

train["GarageFinish_new"]=train["GarageFinish"].apply(garagefinish)
```

```python
def kitchenqual(KitchenQual):
    if KitchenQual=="Ex":
        return 4
    elif KitchenQual=="Gd":
        return 3
    elif KitchenQual=="TA":
        return 2
    elif KitchenQual=="Fa":
        return 1

train["KitchenQual_new"]=train["KitchenQual"].apply(kitchenqual)
```

```python
# dummy değişken oluşturma

train["CentralAir_new"]=train["CentralAir"].replace({"Y":1, "N":0})
```

```python
def extercon(ExterCond):
    if ExterCond=="Ex":
        return 5
    elif ExterCond=="Gd":
        return 4
    elif ExterCond=="TA":
        return 3
    elif ExterCond=="Fa":
        return 2
    elif ExterCond=="Po":
        return 1
    
train["ExterCond_new"]=train["ExterCond"].apply(extercon)
```

```python
Y3=train["SalePrice"]
X3=train[["GarageArea","GrLivArea","OverallQual","OverallCond","PavedDrive_new","GarageQual_new",
        "GarageFinish_new","KitchenQual_new","CentralAir_new","ExterCond_new"]]
```

```python
lrm3=linear_model.LinearRegression()
lrm3.fit(X3,Y3)
```

```python
print("Coefficient:", lrm3.coef_)
print("Bias:", lrm3.intercept_)
```

```python
X3=sm.add_constant(X3)
results=sm.OLS(Y3,X3).fit()
results.summary()
```

Modelin R-square değerinin 0.766 çıkması, özellikler hedef değişkeninin %76.6'sını açıklamaktadır, %23.4'ünü açıklamamıştır.
